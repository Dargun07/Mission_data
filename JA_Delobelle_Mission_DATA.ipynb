```python
# Importation de la bibliotheque necessaire

import pandas as pd
```

```python
# Importation du Fichier CSV et transformation en DataFrame

lien = "https://public.opendatasoft.com/api/explore/v2.1/catalog/datasets/donnees-synop-essentielles-omm/exports/csv?lang=fr&refine=nom_reg%3A%22Hauts-de-France%22&refine=nom%3A%22LILLE-LESQUIN%22&qv1=(date%3A%5B2021-12-31T23%3A00%3A00Z%20TO%202024-12-31T22%3A59%3A59Z%5D)&timezone=Europe%2FBerlin&use_labels=true&delimiter=%3B"
df = pd.read_csv(lien,sep=";")
```
